```python
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/FloZewi/E-commerce-Data-Analysis/refs/heads/master/ecommerce_product_dataset.csv"
data=pd.read_csv(url, encoding ='latin 1')
print(data.head())
```

```python
print(data.isnull().sum())
print(data.describe())

data=data[data['StockQuantity']>0]

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(y='Category',data=data,order=data['Category'].value_counts().index)
plt.title('Diferentes Categorias')
plt.xlabel('Nombre Categoria')
plt.ylabel('Productos')
plt.show()
```

```python
import matplotlib . pyplot as plt
# Histograma para la cantidad de ventas
data.head(5)['ProductName'].hist()
plt.xlabel('PRODUCTOS')
plt.ylabel('VENTAS COMPLETADAS')
plt.title('POPULARIDAD DE VENTAS')
plt.show()
```
